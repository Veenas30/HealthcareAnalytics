**<h4>COVID-19:</h4>**

- The full form of **COVID-19** is **Novel Corona Virus Disease 2019** as it was **first identified** in **December 2019** in **Wuhan, China**, and has since **spread globally**, **resulting** in an **ongoing pandemic**.

- It is an **infectious disease caused by severe acute respiratory syndrome**, also known as (SARS-Cov-2).

- The virus that causes COVID-19 is mainly **transmitted through droplets generated** when an **infected person coughs, sneezes, or exhales**. 

- These **droplets** are **too heavy to hang in the air**, and **quickly fall** on **floors** or **surfaces**.

- **You can be infected** by breathing in the virus **if you are within close proximity** of **someone who has COVID-19**, or by **touching** a **contaminated surface** and then your eyes, nose or mouth.

- **Rest** of the **details** are **present in Pre-Read**.

**Observation:**

### **5.2.1 Access Data**

```
individual_data = pd.read_csv('covid19-in-india-new/IndividualDetails.csv')
print('Data Shape:', individual_data.shape)
individual_data.head(3)
```

### **5.2.2 Analyze Missing Data**

```
null_frame = pd.DataFrame(index = individual_data.columns.values)
null_frame['Null Frequency'] = individual_data.isnull().sum().values
percent = individual_data.isnull().sum().values/individual_data.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100
null_frame.transpose()
```

**Observation:**

- We can see that there are **a lot of missing data present** in this dataset.
- **Instead** of **handling** this **missing information**, we will **continue** our **analysis**.
- **Reason:** Our focus on this case is entirely on analysis rather than model development.

### **5.2.3 Analyze Data Values**

```
print('-----id-----')
print(individual_data['id'].unique())
print('-----government_id-----')
print(individual_data['government_id'].unique())
print('-----diagnosed_date-----')
print(individual_data['diagnosed_date'].unique())
print('-----age-----')
print(individual_data['age'].unique())
print('-----gender-----')
print(individual_data['gender'].unique())
print('-----detected_city-----')
print(individual_data['detected_city'].unique())
print('-----detected_district-----')
print(individual_data['detected_district'].unique())
print('-----nationality-----')
print(individual_data['nationality'].unique())
print('-----current_status-----')
print(individual_data['current_status'].unique())
print('-----status_change_date-----')
print(individual_data['status_change_date'].unique())
print('-----notes-----')
print(individual_data['notes'].unique())
```

**Observation:**

- **age** feature **contains inconsistent data** i.e. it contains a range value **"28-35"**.
- **detected_city** feature **contains inconsitent data** with **" " and "\n" values**.
- **Rest** of the **data** values are **consistent** i.e. the data values are relevant.

**Discrepant values in age Feature**

```
age_discrepancy = individual_data[(individual_data['age'] == '28-35')]
print('Total discripancy Count', age_discrepancy.shape[0])
age_discrepancy
```

**Discrepant values in detected_city Feature**

```
detected_city_discrepancy = individual_data[(individual_data['detected_city'] == '           ') | (individual_data['detected_city'] == '\n')]
print('Total discripancy Count', detected_city_discrepancy.shape[0])
detected_city_discrepancy
```

**Performing Operations**

```
individual_data['age'] = individual_data['age'].replace('28-35', -1)
individual_data['age'] = individual_data['age'].replace(np.nan, -1)

individual_data['detected_city'] = individual_data['detected_city'].replace('           ', 'Not Available')
individual_data['detected_city'] = individual_data['detected_city'].replace('\n', 'Not Available')
```

```
print('Total data points having -1 values:', individual_data[individual_data['age'] == -1].shape[0])
print('Total data points having (Not Available) values', individual_data[individual_data['detected_city'] == 'Not Available'].shape[0])
```

### **5.2.4 Analyze & Handling Data Type**

```
type_frame = pd.DataFrame(index = individual_data.columns.values)
type_frame['Type'] = individual_data.dtypes
type_frame.transpose()
```

**Observation:**

- We can see that **diagnosed_date** , **age** and **status_change_date** features are having **inconsistent data type** i.e. object type.

**Performing Operations:**

```
individual_data['diagnosed_date'] = pd.to_datetime(individual_data['diagnosed_date'], format = '%d/%m/%Y')
individual_data['status_change_date'] = pd.to_datetime(individual_data['status_change_date'], format = '%d/%m/%Y')
individual_data['age'] = individual_data['age'].astype(float)
```

### **5.2.5 Analysis Over Data**

**<h4>Question: What is the proportion of different gender including & excluding missing information?</h4>**

```
figure = plt.figure(figsize = [15, 8])
plt.subplot(1, 2, 1)
colors_list = ['lightcoral', 'lightgreen', 'mediumturquoise']

individual_data['gender'].value_counts(dropna = False).plot(kind = 'pie', figsize = [15, 8], autopct = '%1.1f%%', startangle = 90, shadow = True, 
                                                            labels = None, pctdistance = 1.12, colors = colors_list, explode = np.ones(3)/10, 
                                                            wedgeprops = dict(width = 0.15))
plt.ylabel(ylabel = 'Gender', size = 14)
plt.title('Proportion of each Gender Type (Including NaN)', size = 16)
plt.legend(labels = ['NaN', 'Male', 'Female'], frameon = False)


plt.subplot(1, 2, 2)
colors_list = ['lightcoral', 'lightgreen']

individual_data['gender'].value_counts().plot(kind = 'pie', figsize = [15, 8], autopct = '%1.1f%%', startangle = 90, shadow = True, labels = None, 
                                              pctdistance = 1.12, colors = colors_list, explode = np.ones(2)/10, wedgeprops = dict(width = 0.15))
plt.ylabel(ylabel = 'Gender', size = 14)
plt.title('Proportion of each Gender Type (Excluding NaN)', size = 16)
plt.legend(labels = ['Male', 'Female'], frameon = False)
plt.tight_layout(pad = 3.0)
plt.show()
```

**Observation:**

- **81%** of the **patients gender information** is **missing**. 
- Now if we get rid on missing data, we are **left with** only **19% of consistent information**.
- **Out of this 19%** information there are **~67% males and 33% females**. 
- Let's analyse further with remaining the data.

**<h4>Question : What is the nationality of majority of people inside India?</h4>**

```
# -------------------------------------------------------------------Donut Chart-------------------------------------------------------------------
labels = individual_data['nationality'].value_counts(dropna = False).index
percentage = individual_data['nationality'].value_counts(dropna = False).values/individual_data.shape[0]

colors = []
for i in range(len(individual_data['nationality'].unique())):
  colors.append('#%06X' % randint(0, 0xFFFFFF))

figure = plt.figure(figsize = [15, 7])
plt.subplot(1, 2, 1)
individual_data['nationality'].value_counts(dropna = False).plot(kind = 'pie', figsize = [15, 7], startangle = 90, shadow = True, 
                                                            labels = None, pctdistance = 1.12, colors = colors, 
                                                            explode = np.ones(len(individual_data['nationality'].unique()))/10, 
                                                            wedgeprops = dict(width = 0.15))
plt.legend(labels=['%s (%1.3f %%)' % (l, s) for l, s in zip(labels, percentage*100)], loc = 'center', title = 'Nationality', frameon = False)
plt.ylabel('Nationality', size = 14)

# -------------------------------------------------------------------Bar Chart---------------------------------------------------------------------
colors.pop(0)
plt.subplot(1, 2, 2)
ax = sns.countplot(y = 'nationality', data = individual_data, palette = colors)
total = individual_data.shape[0]
for p in ax.patches:
        percentage = '{:1.3f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width()
        y = p.get_y() + p.get_height() / 2
        ax.annotate(percentage, (x, y))
ax.set_xlabel('Frequency', size = 14)
ax.set_ylabel('Nationality', size = 14)
ax.grid(b = True, axis = 'y')
plt.suptitle('Nationality of Diagnosed People', size = 16, y = 1.02)
plt.tight_layout(pad = 2.0)
plt.show()
```

**Observation:**

- We can see that the **nationality of 90%** of people **was not recorded or disclosed**.
- In the **remaining 10%** there are **9.5% of people** who **are Indian**.

**<h4>Question : What is the current status of different age individuals?</h4>**

```
recovered_status = individual_data[individual_data['current_status'] == 'Recovered']
recovered_status_grouped = recovered_status.groupby(by = 'diagnosed_date', as_index = False).agg('count')

hospitalized_status = individual_data[individual_data['current_status'] == 'Hospitalized']
hospitalized_status_grouped = hospitalized_status.groupby(by = 'diagnosed_date', as_index = False).agg('count')

deceased_status = individual_data[individual_data['current_status'] == 'Deceased']
deceased_status_grouped = deceased_status.groupby(by = 'diagnosed_date', as_index = False).agg('count')

migrated_status = individual_data[individual_data['current_status'] == 'Migrated']
migrated_status_grouped = migrated_status.groupby(by = 'diagnosed_date', as_index = False).agg('count')
```

```
fig = make_subplots(rows = 2, cols = 2, subplot_titles=('Recovered', 'Hospitalized', 'Deceased', 'Migrated'))

fig.add_trace( go.Scatter(x = recovered_status_grouped['diagnosed_date'], y = recovered_status_grouped['age'], mode = 'lines+markers', name = 'Recovered'), row = 1, col = 1 )
fig.add_trace( go.Scatter(x = hospitalized_status_grouped['diagnosed_date'], y = hospitalized_status_grouped['age'], mode = 'lines+markers', name = 'Hospitalized'), row = 1, col = 2 )
fig.add_trace( go.Scatter(x = deceased_status_grouped['diagnosed_date'], y = deceased_status_grouped['age'], mode = 'lines+markers', name = 'Deceased'), row = 2, col = 1 )
fig.add_trace( go.Scatter(x = migrated_status_grouped['diagnosed_date'], y = migrated_status_grouped['age'], mode = 'lines+markers', name = 'Migrated'), row = 2, col = 2 )

fig.update_layout(height = 768, width = 1000, title_text = 'Current Status of Individual', title_x = 0.5, showlegend = False)

# Saving Figure
plot(figure_or_data = fig, filename = 'Current Status of Individual', auto_open = False, image_width = 1280, image_height = 720, image = 'png')
fig.show()
```

**Note:** Migrated is referring to the transporting of person from one place to another.

**Observation:**

- From above it can be seen that **Hospitalized cases** are **more dominating** than **Recovered, Migrated and Deseased cases**.
- If you see **hosiptalized cases on 4 March, 2020** were **pretty low (5)** while **recovery cases were high (17)** with **zero deaths**.
- We can also see that **hospitalized cases** have **increased per day** and were **highest on 25 April, 2020 (1835 cases)**.
- There were **highest number of deaths** on **3 April, 2020**.
- There is only **one case** that has been **migrated**.

**<h4>Question : What is the proportion of current status of individuals?</h4>**

- **Note:** There are high number of cases that are hospitalized. So first we will included hopitalized cases and then exclude hospitalized cases for better clarification.

```
labels = individual_data['current_status'].value_counts().index
values = individual_data['current_status'].value_counts().values

labels_excluding = individual_data[individual_data['current_status'] != 'Hospitalized']['current_status'].value_counts().index
values_excluding = individual_data[individual_data['current_status'] != 'Hospitalized']['current_status'].value_counts().values

fig = make_subplots(rows = 1, cols = 2, specs=[[{"type": "pie"}, {"type": "pie"}]], subplot_titles = ('Including Hospitalized Cases', 'Excluding Hospitalized Cases'))

fig.add_trace( go.Pie(labels = labels, values = values, hole = .8, pull = [0, 0, 0.2]), row = 1, col = 1 )
fig.add_trace( go.Pie(labels = labels_excluding, values = values_excluding, hole = .8, pull = [0, 0, 0.2]), row = 1, col = 2 )

fig.update_layout(height = 600, width = 1000, title_text = 'Proportion of Current Status of Individuals', title_x = 0.5)

# Saving Figure
plot(figure_or_data = fig, filename = 'Proportion of Current Status of Individuals', auto_open = False, image_width = 1280, image_height = 720, image = 'png')
fig.show()
```

**Observation:**

- Yikes! **99% (27,662 cases)** of the people are **hospitalized** while for the **rest** of the people **are either Deceased, Recovered or Migrated**.
- **Out of 27890 cases**, only **181 patients** have been **recovered** while **46 patients** have **died** and **1 was migrated**.

**Observation:**

- **DateTime** feature is having **inconsistent data type**, it **must be changed to datetime format**.

**Performing Operations:**

```
# Convert 13/03/20 18:00 -> 13
icmr_test_data['day'] = icmr_test_data['DateTime'].apply(lambda x: x.split('/')[0])
# Convert 13/03/20 18:00 -> 03
icmr_test_data['month'] = icmr_test_data['DateTime'].apply(lambda x: x.split('/')[1])
# Convert 13/03/20 18:00 -> 20 18:00
icmr_test_data['year'] = icmr_test_data['DateTime'].apply(lambda x: x.split('/')[2])
# Convert 20 18:00 -> 20
icmr_test_data['year'] = icmr_test_data['year'].apply(lambda x: x.split(' ')[0])
# Convert 20 -> 2020
icmr_test_data['year'] = icmr_test_data['year'].replace('20', '2020')

# Adding back to Original Column 13/03/20 18:00 -> 13/03/2020
icmr_test_data['DateTime'] = icmr_test_data['day'].map(str) + '/' + icmr_test_data['month'].map(str) + '/' +icmr_test_data['year'].map(str)

# Transforming to Datetime
icmr_test_data['DateTime'] = pd.to_datetime(icmr_test_data['DateTime'], format = '%d/%m/%Y')

# Dropping day, month and year, SNo column
icmr_test_data.drop(['day', 'month', 'year', 'SNo'], axis = 1, inplace = True)
```

### **5.4.5 Analysis Over Data**

- **Note:** In this section we will **analyze** the **total samples tested, total individual tested, total positive cases per day**.
- In order to do that we need to **uncover** the **cummulative sum** which we have performed below.

**Performing Operations**

```
# Sorting data by latest date
icmr_test_data = icmr_test_data.sort_values(by = 'DateTime', ascending = False).reset_index(drop = True)

# Uncovering the cummulative sum of cases
TotalSamplesTestedperDay = []
TotalIndividualsTestedperDay = []
TotalPositiveCasesperDay = []

size = icmr_test_data.shape[0]
for i, j in zip(range(size), range(1, size)):
  TotalSamplesTestedperDay.append(icmr_test_data.iloc[i, 1] - icmr_test_data.iloc[j, 1])
  TotalIndividualsTestedperDay.append(icmr_test_data.iloc[i, 2] - icmr_test_data.iloc[j, 2])
  TotalPositiveCasesperDay.append(icmr_test_data.iloc[i, 3] - icmr_test_data.iloc[j, 3])

icmr_test_data['TotalSamplesTestedperday'] = pd.Series(TotalSamplesTestedperDay)
icmr_test_data['TotalIndividualsTestedperDay'] = pd.Series(TotalIndividualsTestedperDay)
icmr_test_data['TotalPositiveCasesperDay'] = pd.Series(TotalPositiveCasesperDay)
```

### **5.7.3 Analyze Data Values**

```
print('-----Sno-----')
print(covid19_ind_data['Sno'].unique())
print('-----Date-----')
print(covid19_ind_data['Date'].unique())
print('-----Time-----')
print(covid19_ind_data['Time'].unique())
print('-----State/UnionTerritory	-----')
print(covid19_ind_data['State/UnionTerritory'].unique())
print('-----ConfirmedIndianNational-----')
print(covid19_ind_data['ConfirmedIndianNational'].unique())
print('-----ConfirmedForeignNational-----')
print(covid19_ind_data['ConfirmedForeignNational'].unique())
print('-----Cured-----')
print(covid19_ind_data['Deaths'].unique())
print('-----Deaths-----')
print(covid19_ind_data['Deaths'].unique())
print('-----Confirmed-----')
print(covid19_ind_data['Confirmed'].unique())
```

**Observation:**

- **State** feature **contains** **inconsistent** **values** like **Nagaland# & Jharkhand#**.
- **ConfirmedIndianNational** & **ConfirmedForeignNational** **contains "-" value**, should be **replaced with -1**.
- For **rest** of the values in rest of the **features** looks **relevant**.

**Discrepant values in State Feature**

```
discrepancy_state = covid19_ind_data[(covid19_ind_data['State/UnionTerritory'] == 'Nagaland#') | (covid19_ind_data['State/UnionTerritory'] == 'Jharkhand#')]
print('Total Discripancy Count in State Feature:', discrepancy_state.shape[0])
discrepancy_state
```

**Discrepant values in ConfirmedIndianNational & ConfirmedForeignNational Feature**

```
discrepancy_confirmed = covid19_ind_data[(covid19_ind_data['ConfirmedIndianNational'] == '-') | (covid19_ind_data['ConfirmedForeignNational'] == '-')]
print('Total Discripancy Count in ConfirmedIndianNational & ConfirmedForeignNational Feature:', discrepancy_confirmed.shape[0])
discrepancy_confirmed.head()
```

**Performing Operations**

```
covid19_ind_data['State/UnionTerritory'] = covid19_ind_data['State/UnionTerritory'].replace('Nagaland#', 'Nagaland')
covid19_ind_data['State/UnionTerritory'] = covid19_ind_data['State/UnionTerritory'].replace('Jharkhand#', 'Jharkhand')

covid19_ind_data['ConfirmedIndianNational'] = covid19_ind_data['ConfirmedIndianNational'].replace('-', -1)
covid19_ind_data['ConfirmedIndianNational'] = covid19_ind_data['ConfirmedIndianNational'].astype(np.int64)

covid19_ind_data['ConfirmedForeignNational'] = covid19_ind_data['ConfirmedForeignNational'].replace('-', -1)
covid19_ind_data['ConfirmedForeignNational'] = covid19_ind_data['ConfirmedForeignNational'].astype(np.int64)
```

### **5.7.4 Analyze & Handling Data Type**

```
type_frame = pd.DataFrame(index = covid19_ind_data.columns.values)
type_frame['Type'] = covid19_ind_data.dtypes
type_frame.transpose()
```

**Observation:**

- **Date** feature has **inconsistent data type** that **must** be **transformed** into **datetime format**.

**Performing Operations**

```
# Convert 30/01/20 -> 30
covid19_ind_data['day'] = covid19_ind_data['Date'].apply(lambda x: x.split('/')[0])
# Convert 30/01/20 -> 01
covid19_ind_data['month'] = covid19_ind_data['Date'].apply(lambda x: x.split('/')[1])
# Convert 30/01/20 -> 20
covid19_ind_data['year'] = covid19_ind_data['Date'].apply(lambda x: x.split('/')[2])
# Convert 20 -> 2020
covid19_ind_data['year'] = covid19_ind_data['year'].replace('20', '2020')

# Adding back to Original Column 30/01/20 -> 30/01/2020
covid19_ind_data['Date'] = covid19_ind_data['day'].map(str) + '/' + covid19_ind_data['month'].map(str) + '/' +covid19_ind_data['year'].map(str)

# Transforming to Datetime
covid19_ind_data['Date'] = pd.to_datetime(covid19_ind_data['Date'], format = '%d/%m/%Y')

# Dropping day, month and year and Sno column
covid19_ind_data.drop(['day', 'month', 'year', 'Sno'], axis = 1, inplace = True)
```
