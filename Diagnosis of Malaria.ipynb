## **1. Introduction**

- Malaria is a **life-threatening disease** caused by parasites that are transmitted to people through the **bites** of infected **female Anopheles mosquitoes**.

- Malaria causes symptoms that typically include **fever**, **tiredness**, **vomiting**, and **headaches**.

- In severe cases it can cause **yellow skin**, **seizures**, **coma**, or **death**.

- Symptoms usually begin **ten to fifteen days** after being bitten by an infected mosquito. 
  
  - If not properly treated, people may have **recurrences** of the disease months later.

- It is **preventable** and **curable**.



<center><img src="https://raw.githubusercontent.com/insaid2018/Domain_Case_Studies/master/Healthcare/Images/mosquito.jpg" width="400px"></center>

### **Signs and Symptoms:**

- A malaria infection is generally characterized by the following signs and symptoms:

  - **Fever**
  
  - **Chills**
  
  - **Headache**

  - **Nausea and Vomiting**
  
  - **Muscle Pain and Fatigue**

<br> 

- Other signs and symptoms may include:

  - **Sweating**
  
  - **Chest or Abdominal Pain**
  
  - **Cough**

<br> 

<center><img src="https://raw.githubusercontent.com/insaid2018/Domain_Case_Studies/master/Healthcare/Images/Symptoms_of_Malaria.png" width="400px""></center>

### **Diagnosis:**

- Malaria is typically diagnosed by the **microscopic examination of blood** using **blood films**, or with **antigen-based rapid diagnostic tests**.



### **Prevalence:**

- The disease is widespread in the **tropical** and **subtropical regions** that exist in a broad band around the **equator** including much of **Sub-Saharan Africa**, **Asia**, and **Latin America**.

- In **2016**, there were **216 million cases** of malaria **worldwide** resulting in an estimated **445,000 to 731,000 deaths**. 

- Approximately **90%** of both **cases and deaths** occurred in **Africa**.

- **Rates** of disease have **decreased** from 2000 to 2015 by **37%**, but **increased** from **2014**, during which there were **198 million cases**.

### **Risk Factor:**

- Malaria is commonly associated with **poverty** and has a major **negative effect** on **economic development**.

- In **Africa**, it is estimated to result in **losses of US$12 billion a year** due to **increased healthcare costs**, **lost ability to work**, and **negative effects on tourism**.

## **2. Problem Statement**

- Malaria is caused by **Plasmodium parasites**, which are transmitted via the bites of infected female Anopheles mosquitoes.

- When an infected mosquito bites you, **parasites** are released into your **bloodstream** and dwell in **red blood cells**.

- Since there are several **different species** of Plasmodium parasites and each requires **specific medications**, early and accurate diagnosis of malaria is essential for both rapid and effective disease management.

- **Medical experts** will **inspect** whether the *blood cells are infected with malaria parasites*, and **identify** the **species** of parasites.

**Current Process:**

1. Once a patient is **suspected** to have Malaria in a hospital, his or her **blood samples** are **sent** to a **certified National Centre for Disease Control** (NCDC) **lab** for notification and diagnosis.

2. An experienced **medical laboratory technologist** in the NCDC lab will **conduct** a fast **diagnosis** by *microscopic examination*.

3. **Results** of status of infection, and the **species** of parasites found will be **reported back** to the hospital to initiate prompt and accurate treatment.

4. At the same time **nucleic acid amplification tests** will be performed for **final diagnosis**.

5. Hospitals will receive the **final diagnostic results** from the NCDC lab **after a while**, and adjust their treatment plan accordingly.

### **7.1 Data Augmentation**

- In order to **avoid overfitting**, we need to **expand** our dataset artificially.

  - We can make our existing dataset even **larger**.

  - The idea is to alter the training data with **small transformations** to **reproduce variations** in the images.

  - For example, the image object is not centered, the scale of the image is not the same, the image is rotated, and so on.

- Approaches that **alter** the **training data** in ways that change the array representation while keeping the **label** the **same** are known as **data augmentation** techniques.

- Some popular augmentations people use are **grayscales**, **horizontal flips**, **vertical flips**, **random crops**, **color jitters**, **translations**, **rotations**, and much more.

- By applying just a couple of these transformations to our training data, we can easily double or triple the number of training examples and create a very **robust model**.

<br>

<center><img src="https://raw.githubusercontent.com/insaid2018/Domain_Case_Studies/master/Healthcare/Images/data_augmentation.jpg" width="900px""></center>

- We will use **Keras**'s `ImageDataGenerator` for Data Augmentation.

- Creating an **image generator** both training and test datasets.

- We will only apply the **transformations** to the **training set** and not the test set.

  - This is because we should **not change** the data in the **test dataset**.

- In the test set we will only **divide each pixel value by 255** to **normalize** the pixel values.

  - As a result, all the pixel values will lie between **0 and 1**.

````
train_image_gen = ImageDataGenerator(rotation_range=30,
                                     width_shift_range=0.1,
                                     height_shift_range=0.1,
                                     rescale=1./255,
                                     shear_range=0.1,
                                     zoom_range=0.1,
                                     horizontal_flip=True,
                                     fill_mode='nearest')
````

````
test_image_gen = ImageDataGenerator(rescale=1./255)
````

- **Comparing** an **original** image with its **augmented** version created by the image generator after data augmentation. 

````
parasitic_img = cv2.imread(parasitic_cell)

plt.figure(1, figsize = (16, 7))
plt.subplot(1, 2, 1)
plt.imshow(parasitic_img)
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(train_image_gen.random_transform(parasitic_img))
plt.title('Data Augmented Image')
plt.xticks([]), plt.yticks([])
````

**Observations:**

- By making such **small changes** we can drastically **increase** the **size** of our train set and make a very robust model.

## **8. Model Development & Evaluation**

- In this section, we will be **building** our Machine Learning model and fitting it with the training data.

- We will build a **Convolutional Neural Network** (CNN) which is used for computer vision tasks like image classification etc.

- Then we will **compile** our model specifying the **loss**, **metrics** and the **compiler**.

- After compiling, we will **train** our model on the training dataset.

<a id = Section801></a>
### **8.1 Building the Model**

- We will use Keras's **Sequential API**.

- This allows us to create a Deep Neural Network by **adding** different **layers** one by one in a **sequence**.

- We will use **3 convolutional** layers each followed by a **max pooling** layer and a **dropout** layer.

  - The convolutional layers will **learn** about the various **features** of the image starting from simple features to specific and complex features.

  - The features learned by convolutional layers are **translation invariant** meaning they are learned **irrespective** of their **position** in the image.

- Then we **flatten** the image (**3D to 2D**) and apply a **Dense** layer on it.

- At last, we calculate the **class probabilities** using the **sigmoid** function.

````
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=image_shape))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
   
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))
````

- Let's see the entire model **summary**.

````
model.summary()
````

- The `summary` method allows us to see each **layer** in the model along with the **output** and the **parameters** involved in each layer.

<a id = Section802></a>
### **8.2 Compiling the Model**

- To compile a model we need to specify a few things like:

  - **Optimizer**: Which optimizer to use to calculate the loss function.

    - Here we use **Adam** optimizer because it's **fast and reliable**.

  - **Loss**: The loss value depends on the problem at hand.

    - In our case, it's a **binary classification** problem.

    - So we are using **binary crossentropy** as the loss function.

  - **Metrics**: This parameter specify which metrics to be calculated apart from the loss during the model training process.

    - We have specified **accuracy** because we want to calculate the accuracy of our model for each epoch.

````
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

<a id = Section803></a>
### **8.3 Training the Model**

**Early Stopping Callback**

- This callback will allow us to **stop** the **training automatically**.

- If the **validation loss** doesn't improve for **3 epochs** it will stop the training at that stage. 

````
early_stop = EarlyStopping(monitor='val_loss', patience=3)
````

**Flow From Directory**

- We haven't loaded our dataset images into the **memory** and we will use them directly from thier **directories**.

- We will create a **flow generator** object for both the train and test sets using `flow_from_directory` method of the **image generators**.

- This will allow us to directly **input** the images from our l**ocal system** to our **model** during training and testing.

- To implement `flow_from_directory`, the **input** folder shoud have the following **format**:

```
input/
    class1/
        img1.jpg
        img2.jpg
        ...

    class2/
        img1.jpg
        ...
    ...
```



- Here, the **name** of the **folder** will automatically be picked up as the **class label** for each image in that folder.

- We have already arranged our images in this format from the very beginning, for both the train and test sets.

- Setting up the flow generator to **flow batches** from directories.

- We will use a **batch size** of **64** implying that 64 images will be send in a single batch in each epoch.

- We will keep on sending the images in batches until **all** the **images** are **seen** by the model.

- This will mark the **end** of an **epoch**.

- We will **repeat** the above **process** for **all** the **epochs**.

````
batch_size = 64
````

- While creating the flow generators we are also **resizing** the images to have a **shape** of **130 x 130**.

  - This is specified using the `target_size` parameter, and passing the value **image_shape[:2]**

````
train_generator = train_image_gen.flow_from_directory(train_dir,
                                                      target_size=image_shape[:2],
                                                      color_mode='rgb',
                                                      batch_size=batch_size,
                                                      class_mode='binary')
````

````
test_generator = test_image_gen.flow_from_directory(test_dir,
                                                    target_size=image_shape[:2],
                                                    color_mode='rgb',
                                                    batch_size=batch_size,
                                                    class_mode='binary',
                                                    shuffle=False)
````

- Both the train and test flow generators are **created**.

- Let's see the **target labels**.

````
train_generator.class_indices
````

**Training the Model**

- We will **fit** our model with the **train_generator** and use the **test_generator** for **validation**.

- The training process will be run for **10 epochs**.

- We will also specify the **early stopping** callback during the fitting process.

````
history = model.fit(train_generator, epochs=10, steps_per_epoch=len(train_generator),
                    validation_data=test_generator, validation_steps=len(test_generator), 
                    callbacks=[early_stop])
````

- The model is trained for **10 epochs**.

- **Each epoch** took about **102 seconds**.

- **Total training time** = **1020 seconds** = **17 minutes** 

**Saving the Model**

- We can **save** our **model** for future use.

- This will help us **avoid training** our model again in the future.

````
# Saving the model
model.save('model.h5')
````

- Our model is saved as **model.h5**

**Loading the Model**

- To **load** our saved model, we can use **Keras**'s `load_model`.

````
from tensorflow.keras.models import load_model
````

````
# Loading the saved model as model
model = load_model('model.h5')
````

- Our saved model is loaded as **model**.

<a id = Section804></a>
### **8.4 Evaluating the Model**

- Finally, we will **evaluate** our model's **performance** on both the training and test sets.

- Plotting **training** and **validation accuracy** against **epochs**.

````
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'r', label='Training Accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation Accuracy')
plt.title('Training and Validation Accuracy Plot')
plt.legend()
````

- Plotting **training** and **validation loss** against **epochs**.

````
loss = history.history['loss']
val_loss = history.history['val_loss']
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and Validation Loss Plot')
plt.legend()
````

- **Evaluating** the model performance for the **test set**.

````
model.evaluate(test_generator, steps=len(test_generator))
````

**Observations:**
